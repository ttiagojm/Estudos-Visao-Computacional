```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Coordenadas v√©rtices do cubo
xs = np.array([5, 5, 5, 5, 2, 2, 2, 2])
ys = np.array([3, -3, -3, 3, 3, -3, -3, 3])
zs = np.array([-2, -2, 2, 2, -2, -2, 2, 2])

```

```python
ax = plt.axes(projection='3d')

ax.scatter(xs, ys, zs, color="c")
```

```python
# Coordenadas ortogr√°ficas

orto_mat = np.hstack((np.identity(2), np.zeros((2, 1))))

points_3d = np.vstack((xs, ys, zs))
points = np.zeros((points_3d.shape[0]-1, points_3d.shape[1]))

for col in range(points_3d.shape[1]):
    points[:,col] = orto_mat @ points_3d[:,col] 

orto_xs = points[0, :]
orto_ys = points[1, :]
```

```python
plt.scatter(orto_xs, orto_ys, color="c")
```

```python
# Coordenadas (perspetiva)

proj_xs = xs / zs
proj_ys = ys / zs

plt.scatter(proj_xs, proj_ys, color="c")
```
